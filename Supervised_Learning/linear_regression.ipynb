```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Set seed for random data generation
np.random.seed(0)

# Generate random X values between 0 and 2
X = np.random.rand(100, 1) * 2  
# Generate y values using the equation: y = 3 + 4X + noise
y = 3 + 4 * X + np.random.rand(100, 1)  

# Plot scatter graph of the data
plt.scatter(X, y, label="Actual Data", color="black")

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Plot the model's predictions
plt.plot(X, model.predict(X), color="red", alpha=0.8, label="Linear Regression")

# Plot the function graph using slope (a) and intercept (b) values
for _ in range(100):
    plt.plot(X, model.coef_[0][0] * X + model.intercept_[0], color="blue", alpha=0.1)

plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Linear Regression Graph")
plt.legend()
plt.show()


# ------------------------------------------------------------
# Linear regression analysis on the diabetes dataset

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = load_diabetes(return_X_y=True)

# Use only one feature (variable)
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Split the dataset into training and testing sets
X_train, X_test = diabetes_X[:-20], diabetes_X[-20:]
y_train, y_test = diabetes_y[:-20], diabetes_y[-20:]

# Create and train the linear regression model
regr_model = LinearRegression()
regr_model.fit(X_train, y_train)

# Make predictions using the test data
y_pred = regr_model.predict(X_test)

# Calculate error metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Print the results
print(f"Root Mean Square Error (RMSE): {rmse:.2f}")
print(f"R-Squared Score: {r2:.2f}")
```
