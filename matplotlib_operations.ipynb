```python
import matplotlib.pyplot as plt
```

**Line Plot**

```python
# x ekseni, y ekseni, color, alpha(opacity)
plt.figure(figsize=(6,2))
plt.plot(veri["Id"],veri["SepalLengthCm"], color = "red", alpha = 1, label = "SepalLengthCm") # line plot
plt.plot(veri["Id"],veri["PetalLengthCm"], color = "black", alpha = 1, label = "PetalLengthCm")
plt.title("İris Çiçeği Görselleştirme")
plt.xlabel("Id")
plt.ylabel("SepalLengthCm")
plt.grid(True)
plt.legend()
plt.show()
```

```python
veri.Species.unique()
```

```python

# Korelasyon, iki değişken arasındaki ilişkiyi ölçen bir kavramdır. İki değişkenin nasıl hareket ettiğini gösterir:

# Pozitif Korelasyon: Bir değişken artarken diğeri de artar.
# Negatif Korelasyon: Bir değişken artarken diğeri azalır.
# Korelasyon Yok: Değişkenler arasında ilişki yoktur.
plt.figure(figsize=(6,2))
plt.scatter(veri["SepalLengthCm"], veri["PetalLengthCm"], color = "red", s = 15, alpha = 0.7, label = 1)
plt.title("İris Çiçeği Görselleştirme")
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.grid(True)
plt.legend()
plt.show()
```

```python
veri.Species.unique()
```

```python
veri_setosa = veri[veri.Species == 'Iris-setosa']
veri_versicolor = veri[veri.Species == 'Iris-versicolor']
veri_virginica = veri[veri.Species == 'Iris-virginica']
```

```python
plt.figure(figsize = (6,2))
# scatter(x,y,renk,size, saydamlik)
plt.scatter(veri_setosa["SepalLengthCm"], veri_setosa["PetalLengthCm"], color = "black", s = 15, alpha = 0.7, label = "veri_setosa")
plt.scatter(veri_versicolor["SepalLengthCm"], veri_versicolor["PetalLengthCm"], color = "red", s = 15, alpha = 0.7, label = "veri_versicolor")
plt.scatter(veri_virginica["SepalLengthCm"], veri_virginica["PetalLengthCm"], color = "blue", s = 15, alpha = 0.7, label = "veri_virginica")
plt.title("İris Çiçeği Görselleştirme")
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.grid(True)
plt.legend()
plt.show()
```
