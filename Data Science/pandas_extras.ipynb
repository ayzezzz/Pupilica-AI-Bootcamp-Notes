**List Comprehansion**

```python
import pandas as pd
import numpy as np
```

```python
dictionary = {"isim": ["ali", "veli", "kenan", "murat", "ayse", "hilal"],
              "yas": [15, 16, 17, 33, 45 , 66],
              "sehir": ["izmir", "ankara", "konya", "ankara", "ankara","antalya"]}
veri=pd.DataFrame(dictionary)
veri
```

```python
"""
  Problem Tanımı:

  İnsanların yaşlarının ortalamasını hesapla.
  Yaşları ortalamadan büyük ve küçük olarak grupla.
  Büyük olanlara "büyü", küçük olanlara "küçük" özelliği ekle.
"""

# yas sutununu elde et.
veri_yas = veri.yas
print(veri_yas)

# yas sutununun ortalamasını bul.
veri_yas_mean = np.mean(veri_yas)
print(f"veri_yas_mean: {veri_yas_mean}")
# for döngüsü ile yas ort. hesapla.
# küçük olanlara liste içine küçük, büyük olanlara büyük yazdir.
liste_yas_grubu = []
liste_yas_grubu = []
for i in range(len(veri_yas)):
  dummy_veri_yas = veri_yas[i]
  if veri_yas_mean > dummy_veri_yas:
    liste_yas_grubu.append("kucuk")
  else:
    liste_yas_grubu.append("buyuk")
print(liste_yas_grubu)
veri["yas_grubu"] = liste_yas_grubu
print(veri)
```

```python
veri["yas_grubu2"]=["kucuk" if veri_yas_mean > veri_yas[i] else "buyuk" for i in range(len(veri_yas))]
veri
```

```python
veri["YAS_GRUBU2"]=[i.upper() for i in veri.yas_grubu2]
veri
```

```python
"""
Problem Tanımı:

Tüm satırları kontrol et.
Eğer kişi Ankara'da yaşıyorsa, yaşını 2 ile çarp.
Aksi halde, yaşını 5 ile çarp.
Yeni değerleri tutmak için new_yas adında bir sütun oluştur.
"""
veri["new_yas"]=[veri.yas[i]*2 if veri.sehir[i]=="ankara" else veri.yas[i]*5 for i in range(len(veri.yas))]
veri
```

```python
def yas10YilSonrası(age):
  output = age + 10
  return output
```

```python
# apply fonksiyonu, bir DataFrame veya Series üzerinde özel bir işlemi her bir eleman veya satır/kolona uygulamak için kullanılır.

veri["yas_10_yil_sonrasi"] = veri.yas.apply(yas10YilSonrası)
veri
```

```python
def yasCarpim(df):
  if df.sehir == "ankara":
    return df.yas * 2
  elif df.sehir == "izmir":
    return df.yas * 3
  else:
    return df.yas * 5
```

```python
veri.apply(yasCarpim, axis = 1) # axis=1, bir DataFrame üzerinde satır bazında işlem yapılmasını sağlar
```

```python
"""
Problem Tanımı:

DataFrame'deki insanların isimlerinin uzunluğunu kontrol et:
İsim uzunluğu 5 ise, yaşlarını 2 ile çarp.
İsim uzunluğu 4 ise, yaşlarını 3 ile çarp.
Aksi halde, yaşlarını 4 ile çarp.
Yeni yaşları tutmak için new_yas adında bir sütun oluştur.
"""
def nameLength(df):
  if len(df.isim)==5:
    return df.yas*2
  elif len(df.isim)==4:
    return df.yas*3
  else:
    return df.yas*4
veri["new_age3"]=veri.apply(nameLength, axis=1)
veri
```

**Birleştirme**
*   concat fonksiyonu, iki veya daha fazla DataFrame veya Series'i birleştirmek için kullanılır. Verileri satır bazında ya da sütun bazında birleştirerek yeni bir DataFrame oluşturur.



```python
sozluk1 = {"isim": ["ali", "veli"],
           "yas": [15,16]}
veri1 = pd.DataFrame(sozluk1)
veri1
```

```python
sozluk2 = {"isim": ["ali", "veli"],
           "yas": [32,33]}
veri2 = pd.DataFrame(sozluk2)
veri2
```

```python
# concat
veri_dikey = pd.concat([veri1, veri2], axis = 0)
veri_dikey
```

```python
# concat
veri_yatay = pd.concat([veri1, veri2], axis = 1)
veri_yatay
```
