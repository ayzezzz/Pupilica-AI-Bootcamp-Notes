```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE

# Rakam veri setini ykleme
digits = datasets.load_digits()
X = digits.data
y = digits.target.astype(int)

# t-SNE ile boyut indirgeme (2 bileene)
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap="tab10", alpha=0.7, edgecolors="k")
plt.colorbar(label="Snflar")
plt.title("t-SNE ile Rakam Kmeleme")
plt.xlabel("t-SNE Bileeni 1")
plt.ylabel("t-SNE Bileeni 2")
plt.show()
```
