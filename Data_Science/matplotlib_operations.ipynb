**Line Plot**

```python
# x ekseni, y ekseni, color, alpha(opacity)
plt.figure(figsize=(6,2))
plt.plot(veri["Id"],veri["SepalLengthCm"], color = "red", alpha = 1, label = "SepalLengthCm") # line plot
plt.plot(veri["Id"],veri["PetalLengthCm"], color = "black", alpha = 1, label = "PetalLengthCm")
plt.title("İris Çiçeği Görselleştirme")
plt.xlabel("Id")
plt.ylabel("SepalLengthCm")
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Korelasyon, iki değişken arasındaki ilişkiyi ölçen bir kavramdır. İki değişkenin nasıl hareket ettiğini gösterir:

# Pozitif Korelasyon: Bir değişken artarken diğeri de artar.
# Negatif Korelasyon: Bir değişken artarken diğeri azalır.
# Korelasyon Yok: Değişkenler arasında ilişki yoktur.
plt.figure(figsize=(6,2))
plt.scatter(veri["SepalLengthCm"], veri["PetalLengthCm"], color = "red", s = 15, alpha = 0.7, label = 1)
plt.title("İris Çiçeği Görselleştirme")
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# dataseti indir: https://github.com/prashant-kikani/breast-cancer-detection/blob/master/breast-cancer-data.csv
# class'larin (target variable) diagnosis ayrilmasi
# 2,3 bir subplot cizdirelim
# ilk satirda     M için scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
# ikinici satirda B için scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
```

```python
import pandas as pd
data = pd.read_csv("breast-cancer-data.csv")
data
```

```python
# 'diagnosis' sütununa göre M (Malignant) ve B (Benign) sınıflarını ayırma
m = data[data["diagnosis"] == "M"]
m.reset_index(drop=True, inplace=True)  # 'M' sınıfındaki veriler için index sırasını sıfırlama
b = data[data["diagnosis"] == "B"]
b.reset_index(drop=True, inplace=True)  # 'B' sınıfındaki veriler için index sırasını sıfırlama

# İlk satırda 'M' sınıfı için görselleştirmeler
# Scatter plot: radius_mean vs texture_mean
plt.subplot(2, 3, 1)
plt.scatter(m.radius_mean, m.texture_mean)

# Histogram: 'M' sınıfı için radius_mean'in dağılımı
plt.subplot(2, 3, 2)
plt.hist(m.radius_mean)

# Line plot: 'M' sınıfı için radius_mean'in değişimi
plt.subplot(2, 3, 3)
plt.plot(m.radius_mean)

# İkinci satırda 'B' sınıfı için görselleştirmeler
# Scatter plot: radius_mean vs texture_mean
plt.subplot(2, 3, 4)
plt.scatter(b.radius_mean, b.texture_mean)

# Histogram: 'B' sınıfı için radius_mean'in dağılımı
plt.subplot(2, 3, 5)
plt.hist(b.radius_mean)

# Line plot: 'B' sınıfı için radius_mean'in değişimi
plt.subplot(2, 3, 6)
plt.plot(b.radius_mean)
```
