```python
"""
  Problem Tanımı:
  Bir DataFrame oluşturulacak ve üzerinde aşağıdaki işlemler yapılacaktır:

  1) Yeni sütun: 'maas' sütunu eklenecek.
  2) Yeni sütun: 'yas/maas' oranı hesaplanarak yeni bir sütun eklenecek.
  3) 'yas/maas' sütunu için ortalama değer hesaplanacak ve ekrana yazdırılacak.
  4) Ankara'da yaşayanların ortalama yaşı bulunacak.
  5) Ankara'da yaşayanların içinden en yüksek maaş alan kişi tespit edilecek.

"""
dictionary = {"isim": ["ali", "veli", "kenan", "murat", "ayse", "hilal"],
              "yas": [15, 16, 17, 33, 45 , 66],
              "sehir": ["izmir", "ankara", "konya", "ankara", "ankara","antalya"]}
veri = pd.DataFrame(dictionary)
veri
```

```python
veri["maas"] = [100,200,300,400,500,600]
veri
```

```python
veri["yas_maas"] = veri["yas"] / veri["maas"]
veri
```

```python
veri["ulke"] = ["turkiye"]*len(veri)
veri
```

```python
ankarada_yasayanlar = veri[veri["sehir"] == "ankara"]
ankarada_yasayanlar
```

```python

ankarada_yasayanlar_yas = veri[veri["sehir"] == "ankara"]["yas"]
ankarada_yasayanlar_yas
```

```python
# Ankara'da yaşayanların bilgilerini içeren DataFrame'in indeksini sıfırlar.
# drop=True: Eski indeks sütununu kaldırır.
# inplace=True: Değişikliği mevcut DataFrame üzerinde yapar, yeni bir kopya oluşturmaz.

ankarada_yasayanlar.reset_index(drop=True, inplace=True)
ankarada_yasayanlar
```

```python
"""
  for dongusu kullanarak ankarada yaşayan ve maximum maas alan kişiyi bulma
"""
maks_maas = 0
maks_isim = ""
for i in range(len(ankarada_yasayanlar)):
    if ankarada_yasayanlar["maas"][i] > maks_maas:
        maks_maas = ankarada_yasayanlar["maas"][i]
        maks_isim = ankarada_yasayanlar["isim"][i]

print(maks_isim)
```
