```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# --- IRIS VERİ SETİ VE BOYUT İNDİRME ---
iris = datasets.load_iris()
X, y = iris.data, iris.target
target_names = iris.target_names
colors = ["red", "blue", "green"]

# PCA ile 2D dönüşüm
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# LDA ile 2D dönüşüm
lda = LDA(n_components=2)
X_lda = lda.fit(X, y).transform(X)


```

```python
# --- PCA Sonuçlarını Görselleştirme ---
plt.figure(figsize=(8, 6))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, label=target_name, alpha=0.7, edgecolors="k")
plt.title("PCA ile 2D Görselleştirme")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend()
plt.grid(True)
```

```python
# --- LDA Sonuçlarını Görselleştirme ---
plt.figure(figsize=(8, 6))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], color=color, label=target_name, alpha=0.7, edgecolors="k")
plt.title("LDA ile 2D Görselleştirme")
plt.xlabel("LDA 1")
plt.ylabel("LDA 2")
plt.legend()
plt.grid(True)

plt.show()
```
